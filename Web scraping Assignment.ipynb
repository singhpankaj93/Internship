```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://www.wikipedia.org/')
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
header=[]
for i in soup.find_all('span',class_="other-project-title jsl10n"):
    header.append(i.text)
header
```

```python
sub_header=[]
for i in soup.find_all('span',class_="other-project-tagline jsl10n"):
    sub_header.append(i.text)
sub_header    
    
```

```python
import pandas as pd
df=pd.DataFrame({'Header':header,'Sub-header':sub_header})
df
```

```python
#Q7Write a python program to scrape mentioned details from dineout.co.in and make data frame-
#i) Restaurant name
#ii) Cuisine
#iii) Location
#iv) Ratings
#v) Image URL

from bs4 import BeautifulSoup 
import requests
import pandas as pd

Page=requests.get('https://www.dineout.co.in/delhi-restaurants/welcome-back')
#Page
```

```python
soup=BeautifulSoup(Page.content)
#soup
```

```python
Rest_name=[]
for i in soup.find_all('a',class_="restnt-name ellipsis"):
    Rest_name.append(i.text)
Rest_name

```

```python
cuisine=[]
for i in soup.find_all('div',class_="detail-info"):
    cuisine.append(i.text)
cuisine
```

```python
loc=[]
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    loc.append(i.text)
loc
```

```python
rating=[]
for i in soup.find_all('div',class_="restnt-rating rating-4"):
    rating.append(i.text)
rating
```

```python
image_url=[]
for i in soup.find_all('img',class_="no-img"):
    image_url.append(i['data-src'])
image_url
```

```python
print(len(Rest_name),len(cuisine),len(loc),len(rating),len(image_url))
```

```python
df=pd.DataFrame({'Restaurant_name':Rest_name,'Cuisine':cuisine,'Location':loc,'image':image_url})
df
```

```python
#Q6Write a python program to scrape the details of most downloaded articles from AI in last 90 days.https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles Scrape below mentioned details and make data frame-
#i) Paper Title
#ii) Authors
#iii) Published Date
#iv) Paper URL

#Solution:

!pip install requests
!pip install bs4
from bs4 import BeautifulSoup
import requests

page=requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
page

```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
headings=[]
for i in soup.find_all('h2',class_="sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg"):
    headings.append(i.text)
headings

```

```python
Authors=[]
for i in soup.find_all('span',class_="sc-1w3fpd7-0 dnCnAO"):
    Authors.append(i.text)
Authors    
```

```python
Published_Date=[]
for i in soup.find_all('span',class_="sc-1thf9ly-2 dvggWt"):
    Published_Date.append(i.text)
Published_Date    
                    
```

```python
df=pd.DataFrame({'Titles':headings,'Authors':Authors,'Published Date':Published_Date})
df
    
    
```
